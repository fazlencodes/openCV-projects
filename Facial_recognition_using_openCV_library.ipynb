**Haar Cascade Classifiers** - object detection 
( https://github.com/opencv/opencv/tree/master/data/haarcascades )
is a trained datasets which able to detect objects, for which they have been trained for.


It contains positive and negative images which are used to train the classifier.

  1. positive images: contains images which we want to be identified

  2. negative images: images which does not contain the object we want to detect

```python
import cv2
from google.colab.patches import cv2_imshow

# specify the location of images and haarCascade file
cascadeFile = 'haarcascade_frontalface_default.xml'
imageFile = 'harry.jfif'


# creating the haar cascade object
faceCascade = cv2.CascadeClassifier(cascadeFile)

# read the image
image = cv2.imread(imageFile)

# convert into grayscale, --> faster
# if colored--> slower --> difficult for the algorithm to read
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY) #BGR-> blue, green, red
# bgr, is a standar color format which taken by a camera



####https://stackoverflow.com/questions/36218385/parameters-of-detectmultiscale-in-opencv-using-python

# detect face in an image
# detectMultiScale()--> used for object detection


#firstparameter -> grayscale image

#secondparameter-> scaleFactor  ,default value,1.3
# which specifying how much the image sice is reduced at eaach image scale
# resizing the larger image into small image, making it detectable by the algorithm

# thirdparameter -> minNeighbors
#higher value results in fewer detection but with higher quality
#3~6 is a good value

#fourthparameter ->minSize
#this parameter determines how small size image you want to detect 


faces = faceCascade.detectMultiScale(gray , scaleFactor=1.5 , minNeighbors=3 , minSize=(30,30))

########draw a rectangle around the face###############
#cv2.rectangle(originalimage, startpoint, endpoint, color,thickness)
#startpoint is X,Y 
# endpoint= X+Y
####################################################################

for x,y,w,h in faces:
  cv2.rectangle(image , (x,y), (x+w, y+h) , (0,255,0),2)


#display the final output
cv2_imshow(image)
```
